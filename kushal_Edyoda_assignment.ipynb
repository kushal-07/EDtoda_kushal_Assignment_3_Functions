```python
# Q3 Calculate the Upper and The lower Case

s= 'The quick Brow Fox'
def count_case(k):
    l=0
    u=0
    for i in k:
        if i.isupper():
            u = u+1
            #print('The value of upper',u)
            
        elif i.islower():
            l = l +1
            
            #print('The value of lower',l)
    print('The number of upper case letters in the given string is',l)
    print('The number of lower case letters in the given string is',u)
    
count_case(s)    
```
